```python
arraymeta.head()
```

Split by regions

```python
cb_arraymeta = arraymeta[arraymeta["region.cb"] == 1]
cb_indices = cb_arraymeta.index.tolist()
cb_expression = expression.iloc[cb_indices]
cb_arraymeta.head()

# ancg
ancg_arraymeta = arraymeta[arraymeta["region.ancg"] == 1]
ancg_indices = ancg_arraymeta.index.tolist()
ancg_expression = expression.iloc[ancg_indices]

# dlpfc
dlpfc_arraymeta = arraymeta[arraymeta["region.dlpfc"] == 1]
dlpfc_indices = dlpfc_arraymeta.index.tolist()
dlpfc_expression = expression.iloc[dlpfc_indices]

print("ancg:", ancg_arraymeta.shape, "rows")
print("dlpfc:", dlpfc_arraymeta.shape, "rows")
```

```python
regions = {
    'cb': 'region.cb',
    'ancg': 'region.ancg',
    'dlpfc': 'region.dlpfc',
}

splits = {}
for short, col in regions.items():
    in_region = arraymeta[col] == 1
    is_70plus = arraymeta['age'] >= 70
    idx_70plus = arraymeta[in_region & is_70plus].index
    idx_u70    = arraymeta[in_region & ~is_70plus].index
    splits[(short, '70+')] = pd.concat(
        [arraymeta.loc[idx_70plus], expression.iloc[idx_70plus]], axis=1
    )
    splits[(short, '70-')] = pd.concat(
        [arraymeta.loc[idx_u70], expression.iloc[idx_u70]], axis=1
    )

    print(f"{short:6} 70+: {len(idx_70plus):4d} rows | 70-: {len(idx_u70):4d} rows")
```
